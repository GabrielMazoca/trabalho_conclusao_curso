```python
import pandas as pd
import wget
from zipfile import ZipFile
from google.colab.data_table import DataTable
```

Acessando a base de dados

```python
url_base = 'https://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/DFP/DADOS/'
```

Criando uma lista com o nome dos arquivos

```python
arquivos_zip = []
for ano in range (2010,2023):
  arquivos_zip.append(f'dfp_cia_aberta_{ano}.zip')
```

Fazendo o dowload de todos os arquivos

```python
for arq in arquivos_zip:
  wget.download(url_base+arq)
```

Extraindo os arquivos do zip

```python
for arq in arquivos_zip:
  ZipFile (arq, 'r').extractall ('CVM')
```

Criando arquivos históricos

```python
!mkdir DRE Empresas CVM
```

```python
nomes = ['DRE_con', 'DRE_ind']
for nome in nomes:
  arquivo = pd.DataFrame ()
  for ano in range (2010,2023):
    arquivo = pd.concat ([arquivo, pd.read_csv (f'CVM/dfp_cia_aberta_{nome}_{ano}.csv', sep=';', decimal=',', encoding='ISO-8859-1')])
    arquivo.to_csv(f'DRE Empresas CVM/dfp_cia_aberta_{nome}_2010-2023.csv', index=False)
```
